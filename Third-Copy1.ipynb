```python
import pandas as pd
df = pd.read_csv('train.csv')
```

```python
df.drop(columns = ['PRACTICE_ENTRY_NUMBER', 'PRACTICE_WINNINGS_NUMBER'], axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.CATEGORY_1 = le.fit_transform(df.CATEGORY_1)
df.CATEGORY_2 = le.fit_transform(df.CATEGORY_2)
```

```python
status_check = df.groupby(['UNIQUE_IDENTIFIER'])['STATUS_CHECK'].transform('mean')
category_1 = df.groupby(['UNIQUE_IDENTIFIER'])['CATEGORY_1'].transform('max')
category_2 = df.groupby(['UNIQUE_IDENTIFIER'])['CATEGORY_2'].transform('max')
active_yn = df.groupby(['UNIQUE_IDENTIFIER'])['ACTIVE_YN'].transform('mean')
entry = df.groupby(['UNIQUE_IDENTIFIER'])['ENTRY'].transform('mean')
revenue = df.groupby(['UNIQUE_IDENTIFIER'])['REVENUE'].transform('sum')
winnings_1 = df.groupby(['UNIQUE_IDENTIFIER'])['WINNINGS_1'].transform('sum')
winnings_2 = df.groupby(['UNIQUE_IDENTIFIER'])['WINNINGS_2'].transform('sum')
discount = df.groupby(['UNIQUE_IDENTIFIER'])['DISCOUNT'].transform('sum')
deposit = df.groupby(['UNIQUE_IDENTIFIER'])['DEPOSIT'].transform('sum')
deposit_2 = df.groupby(['UNIQUE_IDENTIFIER'])['DEPOSIT_2'].transform('sum')
withdraw = df.groupby(['UNIQUE_IDENTIFIER'])['WITHDRAW'].transform('sum')
entry_number = df.groupby(['UNIQUE_IDENTIFIER'])['ENTRY_NUMBER'].transform('sum')
winnings_number = df.groupby(['UNIQUE_IDENTIFIER'])['WINNINGS_NUMBER'].transform('sum')
practice_entry = df.groupby(['UNIQUE_IDENTIFIER'])['PRACTICE_ENTRY'].transform('sum')
practice_winnings = df.groupby(['UNIQUE_IDENTIFIER'])['PRACTICE_WINNINGS'].transform('sum')
```

```python
df.drop(columns = ['DEPOSIT_NUMBER', 'WITHDRAW_NUMBER'], axis = 1, inplace = True)
```

```python
df['STATUS_CHECK'] = status_check
df['CATEGORY_1'] = category_1
df['CATEGORY_2'] = category_2
df['ACTIVE_YN'] = active_yn
df['ENTRY'] = entry
df['REVENUE'] = revenue
df['WINNINGS_1'] = winnings_1
df['WINNINGS_2'] = winnings_2
df['DISCOUNT'] = discount
df['DEPOSIT'] = deposit
df['DEPOSIT_2'] = deposit_2
df['WITHDRAW'] = withdraw
df['ENTRY_NUMBER'] = entry_number
df['WINNINGS_NUMBER'] = winnings_number
df['PRACTICE_ENTRY'] = practice_entry
df['PRACTICE_WINNINGS'] = practice_winnings
```

```python
new_df = df.drop_duplicates(subset = ['UNIQUE_IDENTIFIER'])
```

```python
new_df
```

```python
X = new_df.drop(columns = ['UNIQUE_IDENTIFIER', 'SEQUENCE_NO', 'Y1', 'Y2'])
```

```python
y1 = new_df.Y1
y2 = new_df.Y2
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y1_train, y1_valid = train_test_split(X, y1, test_size = 0.3, random_state = 80)
X_train, X_valid, y2_train, y2_valid = train_test_split(X, y2, test_size = 0.3, random_state = 80)
```

```python
#Neural Network Model
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
model1 = Sequential()
model1.add(Dense(17, input_dim=17, kernel_initializer='normal', activation='relu'))
model1.add(Dense(10, kernel_initializer='normal', activation='relu'))
model1.add(Dense(10, kernel_initializer='normal', activation='relu'))
model1.add(Dense(1, kernel_initializer='normal'))
# Compile model
model1.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
# evaluate model
history = model1.fit(X, y1, validation_split=0.2, epochs=20)
```

```python
model2 = Sequential()
model2.add(Dense(17, input_dim=17, kernel_initializer='normal', activation='relu'))
model2.add(Dense(10, kernel_initializer='normal', activation='selu'))
model2.add(Dense(10, kernel_initializer='normal', activation='relu'))
model2.add(Dense(1, kernel_initializer='normal'))
# Compile model
model2.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
# evaluate model
history2 = model2.fit(X, y2, validation_split=0.2, epochs=20)
```
